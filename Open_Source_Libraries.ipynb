```python
import requests #requests is an open source library
re = requests.get("https://bit.ly/yelpthaifood") #.get() is actually a function that belongs to requests
thaifood = re.json() #converts json data into python data structures
print(thaifood)
```

```python
shopee_re = requests.get("https://shopee.sg/api/v4/pdp/get_pc?shop_id=328875964&item_id=5363838012")
shopee_data = shopee_re.json()
print(shopee_data)
```

```python
ft_re = requests.get("https://ft.com")
print(ft_re.content)
```

```python
from bs4 import BeautifulSoup
website = BeautifulSoup(ft_re.content)
for headline in website.select(".headline"):
  print(headline.text)
```

```python
import pandas as pd #brings the pandas library to your code, abbreviates it as pd
thaifood_df = pd.DataFrame(thaifood)
#.DataFrame() is a function that belongs to pandas
#.DataFrame() converts a list of dict to a dataframe
#a dataframe is a table structure by pandas
thaifood_df.head() #.head() prints first 5 rows of a dataframe
```

```python
#1. restaurants with a rating > 3.5
thaifood_df[thaifood_df["rating"] > 3.5]
```

```python
#2. restaurants with distance of less than 2000
thaifood_df[thaifood_df["distance"] < 2000]
```

```python
#3. both conditions 1 and 2

thaifood_df[(thaifood_df["rating"] > 3.5) & (thaifood_df["distance"] < 2000)]
```

```python
morethan35 = thaifood_df[thaifood_df["rating"] > 3.5]
morethan35.to_csv("thaifood35.csv")
```
